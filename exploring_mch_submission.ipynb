```python
import matplotlib.pyplot as plt
import csv

mch = open('mch-2014-2022.csv', 'r', encoding='utf-8')

mch_read = csv.reader(mch)

x_date=[]
ysales_mch=[]
yprice_mch=[]

for line in mch_read:
    if '-' in line[0]:
        x_date.append(line[0])
        try:
            ysales_mch.append(int(line[1]))
            yprice_mch.append(int(line[2]))
        except ValueError:
            ysales_mch.append(0)
            yprice_mch.append(0)
```

```python
fig = plt.figure(figsize=(30,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
ax.plot(x_date, ysales_mch, label='sales')
ax.plot(x_date, yprice_mch, label='price')
ax.legend()
plt.xticks(rotation=90)
plt.title('ManCheonHong',fontsize=20) ## 타이틀 설정
plt.show() 
```
