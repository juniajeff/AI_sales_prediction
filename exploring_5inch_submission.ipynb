```python
import matplotlib.pyplot as plt
import csv
x=[] #date
y1=[] #all
y2=[] #average

f = open('5inch-2017-2022.csv', 'r', encoding='utf-8')
for line in csv.reader(f):
    if '-' in line[0]:
        x.append(line[0])
        try:
            y1.append(int(line[1]))
            y2.append(int(line[2]))
        except ValueError:
            y1.append(0)
            y2.append(0)
f.close()
```

```python
fig = plt.figure(figsize=(16,5)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
ax.plot(x,y1, label='sales')
ax.plot(x,y2, label='price')
ax.legend()
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 45도 회전 
plt.yticks()
plt.title('Monstera 5"',fontsize=20) ## 타이틀 설정
plt.show() 
```
