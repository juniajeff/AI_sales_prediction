```python
import matplotlib.pyplot as plt
import csv

fuego = open('fuego-2013-2022.csv', 'r', encoding='utf-8')

fuego_read = csv.reader(fuego)

x_date=[]
ysales=[]
yprice=[]

for line in fuego_read:
    if '-' in line[0]:
        x_date.append(line[0])
        try:
            ysales.append(int(line[1]))
            yprice.append(int(line[2]))
        except ValueError:
            ysales.append(0)
            yprice.append(0)
```

```python
fig = plt.figure(figsize=(20,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
ax.plot(x_date, ysales, label='sales')
ax.plot(x_date, yprice, label='price')
ax.legend()
plt.xticks(rotation=90) 
plt.title('Trades of Fuego Rose',fontsize=20) ## 타이틀 설정
plt.show() 
```
