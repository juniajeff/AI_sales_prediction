# Orchid

```python
import csv
from tqdm import tqdm

f = open('/content/drive/MyDrive/project_Yangjae/samethingwithphal/2014_2021_phal.csv', 'r', encoding='utf-8')
reader = csv.reader(f)
cultivar=[] #I could assign that cultivar is an empty set.

#각 라인은 컬럼을 나열한 리스트 타입. iterable
for line in tqdm(reader):
    #전에 없던 품종이면 집합에 추가한다.
    if line[2] not in cultivar and line[2] !='품종':
        cultivar.append(line[2])
cultivar.sort()
print(cultivar)
print('총 %d개가 있습니다.' %len(cultivar))
f.close()
```

```python
from collections import Counter
f = open('/content/drive/MyDrive/project_Yangjae/samethingwithphal/2014_2021_phal.csv', 'r', encoding='utf-8')
reader = csv.reader(f)
allcultivar = []
#각 라인은 컬럼을 나열한 리스트 타입. iterable
for line in tqdm(reader):
    #전에 없던 품종이면 집합에 추가한다.
    if line[2] !='품종':
        allcultivar.append(line[2])
assert len(allcultivar) == 18066, '더하는 과정에서 오류가 있습니다'
f.close()
```
